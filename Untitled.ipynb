```python
def piegraph(attr):
    command = "SELECT COUNT(*)as c,"+attr+" FROM Adult WHERE salary LIKE '%<=50K%' GROUP BY "+ attr+" \
            ORDER BY c"
    
    cur.execute(command)
    att = cur.fetchall()
    att_stat = pd.DataFrame.from_records(att,columns = ['count','attribute'])
    plt.pie(att_stat['count'],labels=att_stat['attribute'], autopct='%0.2f%%', shadow=False)
    plt.title(attr)
    plt.savefig(attr+".png")
```

```python
piegraph("sex")
```

```python
piegraph("education")
```

```python
cur.execute("SELECT COUNT(*)as c,education\
             FROM Adult WHERE salary LIKE '%<=50K%' GROUP BY education \
            ORDER BY c \
            LIMIT 10")
att = cur.fetchall()
att_stat = pd.DataFrame.from_records(att,columns = ['count','attribute'])
plt.pie(att_stat['count'],labels=att_stat['attribute'], autopct='%0.2f%%', shadow=False)
plt.title("education")
plt.savefig("education.png")
```
